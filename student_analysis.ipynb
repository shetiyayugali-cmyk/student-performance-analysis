# Student Performance Analysis

This project analyzes the Student Performance dataset using Python. The objective is to clean, analyze, and visualize the data to understand factors affecting student performance.


```python
import pandas as pd
df = pd.read_csv('/student-mat.csv', sep=';')
df.head()
```

# Data Cleaning

In this section, the dataset is inspected for missing values, duplicate records, and data types.

```python
print("Shape of Dataset:", df.shape)
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

# Data Analysis

This section calculates the average final grade, students scoring above 15, correlation between study time and grades, and gender-wise performance.

```python
df = df.drop_duplicates()
print("Duplicates after removal:", df.duplicated().sum())
```

```python
avg_grade = df['G3'].mean()
print("Average Final Grade:", avg_grade)
```

```python
above_15 = df[df['G3'] > 15]

print("Students scoring above 15:")
print(len(above_15))
```

```python
correlation = df['studytime'].corr(df['G3'])

print("Correlation between Study Time and Final Grade:")
print(correlation)
```

```python
gender_avg = df.groupby('sex')['G3'].mean()

print(gender_avg)
```

# Data Visualization

Visualizations help understand grade distribution, study habits, and performance differences between genders.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, edgecolor='black')
plt.title("Distribution of Final Grades")
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['studytime'], df['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

```python
gender_avg = df.groupby('sex')['G3'].mean()

gender_avg.plot(kind='bar')

plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```

# Conclusion

1. The dataset contains 395 records and 33 columns.
2. No missing values or duplicate records were found.
3. The average final grade (G3) is approximately 10.42.
4. A total of 40 students scored above 15.
5. Study time has a very weak positive correlation (0.098) with final grades.
6. Male students performed slightly better on average than female students.

Overall, the dataset indicates that study time alone does not strongly influence final grades, and student performance depends on multiple factors.
